<h3>Exploratory Data Analysis Menggunakan Histogram Visualisasi

```python
#distribusi total buku dengan harga
df['Price'].plot.hist(bins=50, alpha=0.5)
plt.title('Visualisasi distribusi harga buku')
plt.xlabel('Price')
plt.ylabel('total books')
plt.show()
plt.savefig('Output_visualisasi_distribusi harga buku.png')
```

```python
df.hist(bins=10, alpha=0.5, figsize=(16,10))
plt.show()
plt.savefig('Visualisasi distribusi fitur price, reviews, user rating, and year.png')
```

```python
#Visualisasi sebaran data fitur price
import seaborn as sns
sns.distplot(df["Price"], bins=50)
plt.show()
plt.savefig('Visualisasi sebaran data fitur Price.png')
```

```python
#Visualisasi sebaran data fitur Rating
import seaborn as sns
sns.distplot(df["Rating"], bins=50)
plt.show()
plt.savefig('Visualisasi sebaran data fitur Rating.png')
```

```python
#Menampilkan grafik terhadap fitur Year
sns.countplot(df['Year'])
```

```python
#Menampilkan grafik terhadap fitur Genre
sns.countplot(df['Genre'])
```

```python
#Menampilkan grafik nama author yang memiliki reviews paling banyak
author10 = df[['Author', 'Reviews']].drop_duplicates().nlargest(10, columns='Reviews')
author10.plot.bar(x='Author', y='Reviews', rot=70, title= "Top 10 Author")
```

```python
#Menampilkan grafik nama buku yang memiliki riviews paling tinggi
book10 = df[['Name', 'Reviews']].drop_duplicates().nlargest(10, columns='Reviews')
book10.plot.bar(x='Name', y='Reviews', rot=70, title="Top 10 Books")
```

<h3><center>5. Judul buku yang paling banyak menjadi best selling

```python
bestsel10 = df.copy() #--> Duplicate dataframe
bestsel10['freq'] = bestsel10.groupby('Name')['Year'].transform('count') #--> hitung count frequency of event
bestsel10.drop(bestsel10.columns[[1,2,3,4,5,6]], axis=1, inplace=True) #--> Remove unused columns and remove duplicates
bestsel10.drop_duplicates(inplace=True) 
bestsel10.sort_values('freq', ascending=False).head(10)
```

<h3><center>8. Berapa sih rata-rata harga buku berdasarkan genre bukunya ?
   

```python
price = df.copy()
price['Price_mean'] = price.groupby('Genre')['Price'].transform('mean')
price.drop(price.columns[[0,1,2,3,4,5]], axis=1, inplace=True)
price.drop_duplicates(inplace=True)
price
```

<h3><center>9. Siapa ya nama author yang menjadi best selling ?
   

```python
author10 = df.copy() #--> Duplicate dataframe
author10['freq'] = author10.groupby('Author')['Year'].transform('count') #--> Count frequency of event
author10.drop(author10.columns[[0,2,3,4,5,6]], axis=1, inplace=True) #-->Remove unused column 
author10.drop_duplicates(inplace=True)#--> Remove duplicates
author10.sort_values('freq', ascending=False).head(10)#--> Sort besed on the highest / lowest frequncy
```

<h3><center>10. Berapa sih masing-masing rating pada setiap genre-nya?

```python
Rating = df.copy()
Rating['Rating_mean'] = Rating.groupby('Genre')['Rating'].transform('mean')
Rating.drop(Rating.columns[[0,1,2,3,4,5]], axis=1, inplace=True)
Rating.drop_duplicates(inplace=True)
Rating
```
